手動できれいにして消した画像と同じ画像を消す（残った画像以外を消す）
cropを手動で消してmaskとcombinedでcropにある画像以外の画像を消す

```python
import os
import glob

def sync_directories():
    base_dir = "/home/hiromu/energy/data/26_0413_moreclean"
    crop_dir = os.path.join(base_dir, "crop")
    mask_dir = os.path.join(base_dir, "mask")
    combined_dir = os.path.join(base_dir, "combined")

    sub_dirs = ["A", "B", "C"]
    files_to_delete = []

    for sub in sub_dirs:
        # --- mask フォルダの検証 ---
        mask_target_dir = os.path.join(mask_dir, sub)
        if os.path.exists(mask_target_dir):
            # maskディレクトリ内の *_mask.png を取得
            mask_files = glob.glob(os.path.join(mask_target_dir, "*_mask.png"))
            for mask_path in mask_files:
                filename = os.path.basename(mask_path)
                # ファイル名からベース部分を抽出 (例: IMG_6420_mask.png -> IMG_6420)
                base_name = filename.replace("_mask.png", "")
                
                # 対応する crop ファイルのパスを構築
                expected_crop_path = os.path.join(crop_dir, sub, f"{base_name}_crop.png")
                
                # crop画像が存在しない場合、削除リストに追加
                if not os.path.exists(expected_crop_path):
                    files_to_delete.append(mask_path)

        # --- combined フォルダの検証 ---
        combined_target_dir = os.path.join(combined_dir, sub)
        if os.path.exists(combined_target_dir):
            # combinedディレクトリ内の *_combined.png を取得
            combined_files = glob.glob(os.path.join(combined_target_dir, "*_combined.png"))
            for combined_path in combined_files:
                filename = os.path.basename(combined_path)
                # ファイル名からベース部分を抽出 (例: IMG_6420_combined.png -> IMG_6420)
                base_name = filename.replace("_combined.png", "")
                
                # 対応する crop ファイルのパスを構築
                expected_crop_path = os.path.join(crop_dir, sub, f"{base_name}_crop.png")
                
                # crop画像が存在しない場合、削除リストに追加
                if not os.path.exists(expected_crop_path):
                    files_to_delete.append(combined_path)

    # --- 削除の実行と確認 ---
    if not files_to_delete:
        print("削除対象のファイルはありません。すべて同期されています。")
        return

    print(f"基準となる crop 画像が存在しないため、以下の {len(files_to_delete)} 件のファイルを削除します：")
    for f in files_to_delete[:15]: # 最初の15件だけプレビュー表示
        print(f" - {f}")
    if len(files_to_delete) > 15:
        print(f" ... 他 {len(files_to_delete) - 15} 件")
        
    ans = input("\n本当に削除してよろしいですか？ (y/n): ")
    if ans.lower() == 'y':
        for f in files_to_delete:
            os.remove(f)
        print("不要なファイルの削除が完了しました。")
    else:
        print("削除をキャンセルしました。")

if __name__ == "__main__":
    sync_directories()
```
