```python
import os
import cv2
import numpy as np
import glob
from ultralytics import YOLO

# モデルの読み込み
model = YOLO("/home/YOLO/-327_seg/datasets/train2/weights/best.pt")  # yolo11x-seg

# 画像フォルダのベースパス
base_img_folder = "/home/src/yolo/img/"
base_mask_folder = "/home/src/yolo/mask/"

# collage_* フォルダを取得し処理
for folder_name in os.listdir(base_img_folder):
    collage_folder = os.path.join(base_img_folder, folder_name)
    
    # フォルダでない場合はスキップ
    if not os.path.isdir(collage_folder):
        continue

    # 対応するマスク保存フォルダ
    mask_folder = os.path.join(base_mask_folder, folder_name.replace("collage_", "mask_"))
    os.makedirs(mask_folder, exist_ok=True)
    
    # collage_X フォルダ内のすべての .jpg 画像を取得
    image_paths = glob.glob(os.path.join(collage_folder, "*.jpg"))
    
    for img_path in image_paths:
        # 画像を読み込む
        img = cv2.imread(img_path)
        if img is None:
            print(f"画像が読み込めませんでした: {img_path}")
            continue

        # 推論
        results = model.predict(source=img, conf=0.5)
        
        # マスクの取得
        if not results or not hasattr(results[0], 'masks') or results[0].masks is None:
            print(f"マスクが検出されませんでした: {img_path}")
            continue
        
        # 最初のマスクを取得
        mask = results[0].masks.data[0].cpu().numpy()  # CUDA -> CPU に転送して NumPy に変換

        # マスクの保存処理
        mask = (mask * 255).astype(np.uint8)  # 0-1 を 0-255 に変換
        
        # 保存ファイル名の作成
        filename = os.path.basename(img_path).replace(".jpg", ".png")  # 出力はPNGにする
        mask_path = os.path.join(mask_folder, filename)
        
        cv2.imwrite(mask_path, mask)
        print(f"保存: {mask_path}")
```
